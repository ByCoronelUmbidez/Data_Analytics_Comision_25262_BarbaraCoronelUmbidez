### 5. Calidad de Datos: Valores Nulos y Duplicados
<font color='#191970'>Aplicación de la función `calidad()` para contar **nulos**, **filas duplicadas**, y duplicados por **clave primaria** (`id_venta`, `id_cliente`, `id_campanha`). Los resultados se documentan en una tabla resumen.</font>

```python
from IPython.display import display, Markdown

def calidad(df, nombre, clave=None):
    """
    Analiza la calidad del DataFrame y retorna un diccionario con los resultados.
    """
    resultados = {}

    # Aseguramos que el nombre de la clave esté en minúsculas para una mejor compatibilidad
    if clave:
        clave_lower = clave.lower()
    else:
        clave_lower = None

    print("\n" + "="*70)
    print(f"### 📈 CONTROL DE CALIDAD: {nombre}")
    print("="*70)

    # Creamos una copia del DF con nombres de columna en minúsculas para evitar errores
    df_temp = df.copy()
    df_temp.columns = df_temp.columns.str.lower()

    # 1️⃣ Valores Nulos por Columna
    nulos_serie = df.isna().sum()
    nulos_total = nulos_serie.sum()

    nulos_desc = ", ".join([f"{col} ({count})"
                            for col, count in nulos_serie.items() if count > 0])

    if not nulos_desc:
        nulos_desc = "0"

    print("1️⃣ Valores Nulos por Columna:")
    display(nulos_serie.to_frame("nulos"))

    resultados['DataFrame'] = nombre.split()[0]
    resultados['Nulos_Desc'] = nulos_desc
    resultados['Nulos_Total'] = nulos_total

    # 2️⃣ Contar filas duplicadas completas
    dup_rows = df_temp.duplicated(keep=False).sum()
    print(f"\n2️⃣ Filas duplicadas (exactas): {dup_rows}")
    resultados['Duplicados_Filas'] = dup_rows

    # 3️⃣ Análisis de Duplicados por Columna Clave
    dup_key_count = 0
    if clave_lower and clave_lower in df_temp.columns:
        print("\n3️⃣ Análisis de Duplicados por Columna Clave:")
        dup_key_count = df_temp[clave_lower].duplicated(keep=False).sum()
        print(f"Duplicados por clave '{clave}': {dup_key_count}")

        if dup_key_count > 0:
            # Mostramos el top de duplicados
            duplicados_ordenados = df_temp[df_temp[clave_lower].duplicated(keep=False)][clave_lower].value_counts()
            print("\n🔁 Top valores duplicados más frecuentes:")
            display(duplicados_ordenados.head(3).to_frame(name='Recuento'))

    resultados['Duplicados_Clave'] = dup_key_count

    print("="*70)

    return resultados

# ----------------------------------------------------------------------
```

```python
# ============================================
# ⚙️ APLICACIÓN Y GENERACIÓN DE TABLA RESUMEN
# ============================================

lista_resultados = []

# 1. Análisis del DataFrame 'ventas'
lista_resultados.append(calidad(ventas, "VENTAS (ventas.csv)", clave="id_venta"))

# 2. Análisis del DataFrame 'clientes'
lista_resultados.append(calidad(clientes, "CLIENTES (clientes.csv)", clave="id_cliente"))

# 3. Análisis del DataFrame 'marketing'
lista_resultados.append(calidad(marketing, "MARKETING (marketing.csv)", clave="id_campanha"))


# ============================================
# 📝 GENERACIÓN AUTOMÁTICA DE LA TABLA RESUMEN
# ============================================

# Convertir la lista de resultados en un DataFrame para facilitar la presentación
df_resumen = pd.DataFrame(lista_resultados)

print("\n\n" + "+"*100)
print("🏆 TABLA RESUMEN DE CALIDAD DE DATOS (DOCUMENTACIÓN FINAL) 🏆")
print("+"*100)

# Formato Markdown de la tabla
markdown_table = (
    "| DataFrame | Valores Nulos (Columnas afectadas) | Filas Duplicadas (Conteo) | Duplicados por Clave (Conteo) |\n"
    "| :--- | :--- | :--- | :--- |\n"
)

for index, row in df_resumen.iterrows():
    markdown_table += (
        f"| **{row['DataFrame']}** "
        f"| {row['Nulos_Desc']} "
        f"| {row['Duplicados_Filas']} "
        f"| {row['Duplicados_Clave']} |\n"
    )

# Usamos display(Markdown()) para que Colab renderice la tabla correctamente
display(Markdown(markdown_table))
```

## 🧼 ETAPA 2: Preprocesamiento y Limpieza de Datos

### 1. Limpieza Profunda y Normalización de Datos
<font color='#B88484'>Se ejecuta el proceso de limpieza y normalización: **eliminación de duplicados**, **conversión de tipos de datos** (ej., precio de string a float), **imputación de valores nulos** (usando la mediana o valores fijos), y **homogeneización de texto** (minúsculas y `strip`).</font>

1. Limpieza de Datos: Eliminar Duplicados

```python
print("--- 1. Limpieza de Nulos y Caracteres ---")

# ============================================
# 🧹 LIMPIEZA Y NORMALIZACIÓN DE LOS DATASETS
# ============================================
# Se limpian y normalizan los DataFrames:
#   ventas, clientes, marketing
# ============================================

# -------------------------------------------------
# 1️⃣ Crear copias independientes para no modificar los originales
# -------------------------------------------------
ventas_clean = ventas.copy()
clientes_clean = clientes.copy()
marketing_clean = marketing.copy()

# -------------------------------------------------
# 2️⃣ Eliminar filas completamente duplicadas
# -------------------------------------------------
ventas_clean = ventas_clean.drop_duplicates()
clientes_clean = clientes_clean.drop_duplicates()
marketing_clean = marketing_clean.drop_duplicates()
```

```python
# -------------------------------------------------
# 3️⃣ Limpieza de Tipos de Datos y Conversión
# -------------------------------------------------

# Columna 'precio' en ventas: Quitar el '$' y convertir a float
ventas_clean['precio'] = ventas_clean['precio'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
print("✅ Columna 'precio' de ventas convertida a float.")

# Columna 'fecha_venta' en ventas: Convertir a datetime
ventas_clean['fecha_venta'] = pd.to_datetime(ventas_clean['fecha_venta'], format='%d/%m/%Y', errors='coerce')
print("✅ Columna 'fecha_venta' convertida a datetime.")

# Columna 'ingresos' en clientes: Quitar el '$' y convertir a float (Se fuerza la columna a tipo string ANTES de usar .str)
try:
    clientes_clean['ingresos'] = (
        clientes_clean['ingresos']
        .astype(str) # Paso clave: garantiza que es string para usar .str
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .astype(float)
    )
    print("✅ Columna 'ingresos' de clientes convertida a float.")
except Exception as e:
    print(f"⚠️ Error al limpiar ingresos de clientes: {e}")
    # Si la columna ya estaba como float, simplemente aseguramos el tipo
    clientes_clean['ingresos'] = pd.to_numeric(clientes_clean['ingresos'], errors='coerce')
    print("✅ Columna 'ingresos' de clientes asegurada como numérica.")


# Columna 'costo' en marketing: Convertir a float
# Asumimos que aquí no hay '$', solo nulos o texto que debe ser numérico.
marketing_clean['costo'] = pd.to_numeric(marketing_clean['costo'], errors='coerce')
print("✅ Columna 'costo' de marketing asegurada como numérica.")
```

```python
# -----------------------------------
# 4️⃣ Imputación de Valores Nulos
# -----------------------------------

# En ventas_clean:
# 'cantidad' y 'precio': Rellenamos con la mediana
ventas_clean['cantidad'] = ventas_clean['cantidad'].fillna(ventas_clean['cantidad'].median())
ventas_clean['precio'] = ventas_clean['precio'].fillna(ventas_clean['precio'].median())

# En categóricos de ventas (producto, categoria): Rellenar con 'desconocido'
ventas_clean['producto'] = ventas_clean['producto'].fillna('desconocido')
ventas_clean['categoria'] = ventas_clean['categoria'].fillna('sin categoria')

# En clientes_clean:
# 'edad' e 'ingresos': Rellenamos con la mediana
clientes_clean['edad'] = clientes_clean['edad'].fillna(clientes_clean['edad'].median())
clientes_clean['ingresos'] = clientes_clean['ingresos'].fillna(clientes_clean['ingresos'].median())
# 'ciudad': Rellenar con 'no especificada'
clientes_clean['ciudad'] = clientes_clean['ciudad'].fillna('no especificada')

# En marketing_clean:
# 'costo': Rellenamos con la mediana
marketing_clean['costo'] = marketing_clean['costo'].fillna(marketing_clean['costo'].median())
# 'canal': Rellenamos con un valor fijo
marketing_clean['canal'] = marketing_clean['canal'].fillna('otro')

print("✅ Todos los valores nulos han sido tratados por imputación.")
```

```python
# -------------------------------------------------
# 5️⃣ Homogeneización de Texto (Minúsculas y Strip)
# -------------------------------------------------

# Función auxiliar para aplicar minúsculas y strip a todas las columnas object
def homogeneizar_texto(df):
    for col in df.select_dtypes(include='object').columns:
        # Usamos .astype(str) para prevenir errores si la columna no fuera estrictamente 'object'
        df[col] = df[col].astype(str).str.strip().str.lower()
    return df

ventas_clean = homogeneizar_texto(ventas_clean)
clientes_clean = homogeneizar_texto(clientes_clean)
marketing_clean = homogeneizar_texto(marketing_clean)

print("✅ Columnas de texto estandarizadas a minúsculas y sin espacios extra.")

# Documentación del proceso

print("Se eliminaron duplicados, se normalizaron tipos de datos y se imputaron nulos.")
```

### 2. Transformación: Aplicar filtros y transformaciones para crear una tabla de ventas que muestre solo los productos con alto rendimiento.✨
<font color='#B88484'>Se crea la métrica clave **'ingresos'** (`cantidad * precio`) y se aplica un filtro para generar un nuevo DataFrame (`ventas_alto_rendimiento`) que solo contiene transacciones que superan el **Percentil 75** de ingresos totales.</font>

```python
print("\n--- 2. Transformación: Productos Alto Rendimiento ---")

# A. Calcular Ingresos por transacción (Ingresos = Cantidad * Precio)
# Usamos las columnas numéricas ya limpias
ventas_clean['ingresos'] = ventas_clean['cantidad'] * ventas_clean['precio']
print("Columna 'ingresos' calculada.")

# B. Definir y Filtrar Alto Rendimiento (Top 25% de Ingresos)
# 1. Calcular el umbral (Percentil 75)
umbral_alto_rendimiento = ventas_clean['ingresos'].quantile(0.75)

# 2. Aplicar el filtro para crear la tabla de alto rendimiento
ventas_alto_rendimiento = ventas_clean[
    ventas_clean['ingresos'] >= umbral_alto_rendimiento
].copy() # Usar .copy() para crear una tabla independiente

print(f"Umbral de Alto Rendimiento (P75): ${umbral_alto_rendimiento:.2f}")
print(f"Tabla 'ventas_alto_rendimiento' creada con {len(ventas_alto_rendimiento)} transacciones (Top 25% de ingresos).")

display(ventas_alto_rendimiento[['id_venta', 'producto', 'ingresos']].head(3))
```

### 3. Agregación: Resumir las ventas por categoría de producto y analizar los ingresos generados. 📊
<font color='#B88484'>Se agrupan los datos por la columna **'producto'** para obtener métricas clave agregadas, lo que permite visualizar la contribución de cada producto a los ingresos totales y al volumen de unidades vendidas.</font>

```python
print("\n--- 3. Agregación: Ventas por Categoría e Ingresos ---")

# Agrupar por la columna 'producto' y agregar las métricas
resumen_por_categoria = (
    ventas_clean
    .groupby('producto')
    .agg(
        Total_Unidades=('cantidad', 'sum'),
        Ingresos_Generados=('ingresos', 'sum'),
        Precio_Promedio=('precio', 'mean')
    )
    .sort_values(by='Ingresos_Generados', ascending=False) # Ordenar por los ingresos más altos
)

print("Resumen de ventas por Producto (Top 5 en Ingresos):")
display(resumen_por_categoria.head(5))
```

### 4. Integración de Datos: Ventas y Marketing 🔗
<font color='#B88484'>Se combinan los DataFrames **ventas** y **marketing** usando la clave común **'producto'**. Esto permite analizar si las campañas específicas para un producto (canal, costo) tienen correlación con los ingresos generados por ese mismo producto.</font>

```python
print("\n--- 4. Integración de Datos: Ventas y Marketing (FINAL con KPIs) ---")

# 1. Normalizar nombres de columnas a minúsculas por seguridad
ventas_clean.columns = ventas_clean.columns.str.lower()
marketing_clean.columns = marketing_clean.columns.str.lower()

# 2. Unión Ventas y Marketing (usando 'producto' como clave)
integracion_final = pd.merge(
    ventas_clean,
    marketing_clean[['producto', 'canal', 'costo', 'id_campanha']],
    on='producto',
    how='left',
    suffixes=('_Venta', '_Marketing')
)

print(f"✅ Integración finalizada. DataFrame 'integracion_final' creado con {len(integracion_final)} filas.")
print("\nPrimeras filas del DataFrame unificado:")
display(integracion_final[['id_venta', 'producto', 'ingresos', 'id_campanha', 'canal']].head(5))

# ================================================
# CÁLCULO DE KPIS ADICIONALES SOLICITADOS
# ================================================

# 3. Ingreso total por campaña (id_campanha)
ingreso_por_campanha = (
    integracion_final
    .groupby('id_campanha')['ingresos']
    .sum()
    .sort_values(ascending=False)
    .to_frame('Ingreso Total')
)

print("\nIngreso total por campaña (id_campanha):")
display(ingreso_por_campanha.head(5))

# 4. Ingreso total por canal (canal)
ingreso_por_canal = (
    integracion_final
    .groupby('canal')['ingresos']
    .sum()
    .sort_values(ascending=False)
    .to_frame('Ingreso Total')
)

print("\nIngreso total por canal (canal):")
display(ingreso_por_canal.head(5))
```
